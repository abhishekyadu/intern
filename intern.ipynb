```python
import requests
import pandas as pd

# CoinGecko API URL to fetch the top 50 cryptocurrencies
url = "https://api.coingecko.com/api/v3/coins/markets"
params = {
    'vs_currency': 'usd',
    'order': 'market_cap_desc',
    'per_page': 50,
    'page': 1
}

response = requests.get(url, params=params)
data = response.json()

# Prepare the data in a pandas DataFrame
df = pd.DataFrame(data)
df = df[['name', 'symbol', 'current_price', 'market_cap', 'total_volume', 'price_change_percentage_24h']]

# Display the DataFrame
print(df)
```

```python
top_5 = df.nlargest(5, 'market_cap')
print("Top 5 Cryptocurrencies by Market Cap:\n", top_5[['name', 'market_cap']])
```

```python
avg_price = df['current_price'].mean()
print(f"Average Price of the Top 50 Cryptocurrencies: ${avg_price:.2f}")
highest_change = df.nlargest(1, 'price_change_percentage_24h')
lowest_change = df.nsmallest(1, 'price_change_percentage_24h')
```

```python
print("Highest 24-hour Price Change:\n", highest_change[['name', 'price_change_percentage_24h']])
print("Lowest 24-hour Price Change:\n", lowest_change[['name', 'price_change_percentage_24h']])
```

```python
import openpyxl
from openpyxl import Workbook

# Load or create the workbook
wb = Workbook()
ws = wb.active
ws.title = "Crypto Data"

# Adding headers
headers = ['Name', 'Symbol', 'Current Price (USD)', 'Market Cap', '24h Volume', '24h Change (%)']
ws.append(headers)

# Adding fetched data
for index, row in df.iterrows():
    ws.append([row['name'], row['symbol'], row['current_price'], row['market_cap'], row['total_volume'], row['price_change_percentage_24h']])

# Save the workbook
wb.save("crypto_data.xlsx")
```
